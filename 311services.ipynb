```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure plots display in the notebook
%matplotlib inline
```

```python
# Load the Excel datasets for 2023 and 2024
file_path_2023 = '/content/3-1-1-service-requests 2023.xlsx'
file_path_2024 = '/content/3-1-1-service-requests 2024.xlsx'

# Load the Excel files into pandas DataFrames
df_2023 = pd.read_excel(file_path_2023)
df_2024 = pd.read_excel(file_path_2024)

# Combine the datasets into one DataFrame
df = pd.concat([df_2023, df_2024], ignore_index=True)

# Preview the combined dataset
df.head()
```

```python
# Check the columns of the 2023 dataset
print("Columns in 2023 Dataset:")
print(df_2023.columns)

# Check the columns of the 2024 dataset
print("\nColumns in 2024 Dataset:")
print(df_2024.columns)
```

```python
# Check the data types of the 2023 dataset
print("\nData types of 2023 Dataset:")
print(df_2023.dtypes)

# Check the data types of the 2024 dataset
print("\nData types of 2024 Dataset:")
print(df_2024.dtypes)
```

```python
# Check for missing values in the 2023 dataset
print("\nMissing values in 2023 Dataset:")
print(df_2023.isnull().sum())

# Check for missing values in the 2024 dataset
print("\nMissing values in 2024 Dataset:")
print(df_2024.isnull().sum())
```

```python
# Summary statistics for the 2023 dataset
print("\nSummary statistics for 2023 Dataset:")
print(df_2023.describe(include='all'))

# Summary statistics for the 2024 dataset
print("\nSummary statistics for 2024 Dataset:")
print(df_2024.describe(include='all'))
```

```python
# Compare columns between the two datasets
columns_2023 = set(df_2023.columns)
columns_2024 = set(df_2024.columns)

if columns_2023 == columns_2024:
    print("\nBoth datasets have the same columns.")
else:
    print("\nColumns are different between the datasets:")
    print("Columns in 2023 but not in 2024:", columns_2023 - columns_2024)
    print("Columns in 2024 but not in 2023:", columns_2024 - columns_2023)
```

```python
## Handle missing values
# Option 1: Drop rows with missing values
df_2023_clean = df_2023.dropna(subset=['Service request open timestamp', 'Latitude', 'Longitude'])  # Keep essential columns
df_2024_clean = df_2024.dropna(subset=['Service request open timestamp', 'Latitude', 'Longitude'])  # Keep essential columns

# Option 2: Alternatively, you can fill missing values for specific columns (e.g., fill with a default value)
# df_2023_clean['Latitude'] = df_2023_clean['Latitude'].fillna(0)
# df_2024_clean['Latitude'] = df_2024_clean['Latitude'].fillna(0)

# Step 3: Ensure correct data types
# Convert 'Service request open timestamp', 'Service request close date', and 'Last modified timestamp' to datetime format
df_2023_clean['Service request open timestamp'] = pd.to_datetime(df_2023_clean['Service request open timestamp'], errors='coerce')
df_2023_clean['Service request close date'] = pd.to_datetime(df_2023_clean['Service request close date'], errors='coerce')
df_2023_clean['Last modified timestamp'] = pd.to_datetime(df_2023_clean['Last modified timestamp'], errors='coerce')

df_2024_clean['Service request open timestamp'] = pd.to_datetime(df_2024_clean['Service request open timestamp'], errors='coerce')
df_2024_clean['Service request close date'] = pd.to_datetime(df_2024_clean['Service request close date'], errors='coerce')
df_2024_clean['Last modified timestamp'] = pd.to_datetime(df_2024_clean['Last modified timestamp'], errors='coerce')

# Ensure Latitude and Longitude are numeric (these should already be float64, but we check for invalid values)
df_2023_clean['Latitude'] = pd.to_numeric(df_2023_clean['Latitude'], errors='coerce')
df_2023_clean['Longitude'] = pd.to_numeric(df_2023_clean['Longitude'], errors='coerce')

df_2024_clean['Latitude'] = pd.to_numeric(df_2024_clean['Latitude'], errors='coerce')
df_2024_clean['Longitude'] = pd.to_numeric(df_2024_clean['Longitude'], errors='coerce')

# Step 4: Remove duplicate rows if any
df_2023_clean = df_2023_clean.drop_duplicates()
df_2024_clean = df_2024_clean.drop_duplicates()

# Step 5: Combine the cleaned datasets for 2023 and 2024
df_clean = pd.concat([df_2023_clean, df_2024_clean], ignore_index=True)

# Step 6: Final check of the cleaned data
print("Cleaned dataset columns and data types:")
print(df_clean.dtypes)

# Check for remaining missing values after cleaning
print("\nRemaining missing values in the cleaned dataset:")
print(df_clean.isnull().sum())

# Display first few rows of the cleaned data
df_clean.head()
```

```python
# Total number of service requests
total_requests = df_clean.shape[0]
print(f"Total number of service requests: {total_requests}")
```

```python
# 1. Total number of service requests
total_requests = df_clean.shape[0]
print(f"Total number of service requests: {total_requests}")

# 2. Service requests by department
requests_by_department = df_clean['Department'].value_counts()
print("\nService requests by department:")
print(requests_by_department)

# 3. Service requests by status
requests_by_status = df_clean['Status'].value_counts()
print("\nService requests by status:")
print(requests_by_status)

# Convert 'Service request open timestamp', 'Service request close date' to timezone-naive datetime
df_clean['Service request open timestamp'] = pd.to_datetime(df_clean['Service request open timestamp'], errors='coerce').dt.tz_localize(None)
df_clean['Service request close date'] = pd.to_datetime(df_clean['Service request close date'], errors='coerce').dt.tz_localize(None)

# Now calculate the time to close a service request
df_clean['Time to Close'] = df_clean['Service request close date'] - df_clean['Service request open timestamp']

# Calculate the average time to close a service request
average_time_to_close = df_clean['Time to Close'].mean()
print(f"\nAverage time to close a service request: {average_time_to_close}")


# 5. Service requests by local area (neighborhood)
requests_by_local_area = df_clean['Local area'].value_counts()
print("\nService requests by local area:")
print(requests_by_local_area)

# 6. Service requests by channel
requests_by_channel = df_clean['Channel'].value_counts()
print("\nService requests by channel:")
print(requests_by_channel)

# 7. Average and median latitude and longitude
average_latitude = df_clean['Latitude'].mean()
average_longitude = df_clean['Longitude'].mean()
print(f"\nAverage Latitude: {average_latitude}")
print(f"Average Longitude: {average_longitude}")

median_latitude = df_clean['Latitude'].median()
median_longitude = df_clean['Longitude'].median()
print(f"Median Latitude: {median_latitude}")
print(f"Median Longitude: {median_longitude}")

# 8. Service requests by month
df_clean['Request Month'] = df_clean['Service request open timestamp'].dt.month
requests_by_month = df_clean.groupby('Request Month').size()
print("\nService requests by month:")
print(requests_by_month)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar plot for service requests by department
plt.figure(figsize=(8, 6))
requests_by_department.plot(kind='bar', color='skyblue')
plt.title('Service Requests by Department')
plt.xlabel('Department')
plt.ylabel('Number of Requests')
plt.xticks(rotation=45)
plt.show()
```

```python
# Pie chart for service request status
plt.figure(figsize=(6, 6))
requests_by_status.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightgreen', 'lightcoral'])
plt.title('Service Request Status')
plt.ylabel('')  # Remove the y-label for clarity
plt.show()
```

```python
# Bar plot for service requests by local area
plt.figure(figsize=(10, 6))
requests_by_local_area.plot(kind='bar', color='lightblue')
plt.title('Service Requests by Local Area')
plt.xlabel('Local Area (Neighborhood)')
plt.ylabel('Number of Requests')
plt.xticks(rotation=90)
plt.show()
```

```python
# Separate the cleaned data into 2023 and 2024 datasets
df_2023 = df_clean[df_clean['Service request open timestamp'].dt.year == 2023]
df_2024 = df_clean[df_clean['Service request open timestamp'].dt.year == 2024]

# Check if the separation worked
print(f"2023 dataset shape: {df_2023.shape}")
print(f"2024 dataset shape: {df_2024.shape}")
```

```python
# For 2023 dataset

# 1. Bar Plot for Service Requests by Department
plt.figure(figsize=(8, 6))
df_2023['Department'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Service Requests by Department (2023)')
plt.xlabel('Department')
plt.ylabel('Number of Requests')
plt.xticks(rotation=45)
plt.show()

# 2. Pie Chart for Service Request Status (2023)
plt.figure(figsize=(6, 6))
df_2023['Status'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightgreen', 'lightcoral'])
plt.title('Service Request Status (2023)')
plt.ylabel('')
plt.show()

# # 3. Bar Plot for Service Requests by Local Area (2023)
plt.figure(figsize=(10, 6))
df_2023['Local area'].value_counts().plot(kind='bar', color='lightblue')
plt.title('Service Requests by Local Area (2023)')
plt.xlabel('Local Area (Neighborhood)')
plt.ylabel('Number of Requests')
plt.xticks(rotation=90)
plt.show()

# # 4. Pie Chart for Service Requests by Channel (2023)
plt.figure(figsize=(6, 6))
df_2023['Channel'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral', 'lightgreen', 'orange'])
plt.title('Service Requests by Channel (2023)')
plt.ylabel('')
plt.show()

# # 5. Histogram of Time to Close a Service Request (2023)
plt.figure(figsize=(8, 6))
df_2023['Time to Close (days)'] = df_2023['Time to Close'].dt.total_seconds() / (24 * 3600)
plt.hist(df_2023['Time to Close (days)'].dropna(), bins=30, color='lightgreen', edgecolor='black')
plt.title('Distribution of Time to Close Service Requests (2023)')
plt.xlabel('Time to Close (Days)')
plt.ylabel('Number of Requests')
plt.show()
```
